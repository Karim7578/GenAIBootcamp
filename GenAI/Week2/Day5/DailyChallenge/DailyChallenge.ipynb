```python
import math

class Cercle:
    def __init__(self, rayon=None, diametre=None):

        if rayon is not None:
            self.rayon = rayon
        elif diametre is not None:
            self.rayon = diametre / 2
        else:
            raise ValueError("Must specify either radius or diameter")
    
    @property
    def diametre(self):
        return self.rayon * 2
    
    def aire(self):
        return math.pi * self.rayon ** 2
    
    # Méthodes spéciales (dunder methods)
    def __str__(self):
        return f"Cercle de rayon {self.rayon}, diamètre {self.diametre}, aire {self.aire():.2f}"
    
    def __add__(self, autre):
        return Cercle(rayon=self.rayon + autre.rayon)
    
    def __eq__(self, autre)
        return self.rayon == autre.rayon
    
    def __lt__(self, autre):
        return self.rayon < autre.rayon

#Test
if __name__ == "__main__":
    c1 = Cercle(rayon=5)
    c2 = Cercle(diametre=10)
    c3 = Cercle(rayon=7)
    
    print("Cercle 1:", c1)
    print("Cercle 3:", c3)
    
    c4 = c1 + c3
    print("\nAddition de c1 et c3:")
    print(c4)
    
    print("\nComparaisons:")
    print("c1 == c2?", c1 == c2)  # True
    print("c1 < c3?", c1 < c3)    # True
    
    liste_cercles = [c3, c1, Cercle(rayon=2)]
    cercles_tries = sorted(liste_cercles)
    
    print("\nCercles triés du plus petit au plus grand:")
    for cercle in cercles_tries:
        print(cercle)
```
