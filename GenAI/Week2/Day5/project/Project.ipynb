```python
import random

class Game:
    def get_user_item(self):
        while True:
            user_input = input("Choisissez un élément (pierre/feuille/ciseaux) : ").lower()
            if user_input in ['pierre', 'feuille', 'ciseaux']:
                return user_input
            print("Choix invalide. Veuillez choisir pierre, feuille ou ciseaux.")

    def get_computer_item(self):
        computer_item = random.randint(1, 3)
        if computer_item == 1:
            return "pierre"
        elif computer_item == 2:
            return "feuille"
        else:
            return "ciseaux"
        
    def get_game_result(self, user_item, computer_item):
        if user_item == computer_item:
            return 'égalité'
        winning_combinations = [('pierre', 'ciseaux'), ('feuille', 'pierre'), ('ciseaux', 'feuille')]
        if (user_item, computer_item) in winning_combinations:
            return 'gagné'
        return 'perdu'

    def play(self):
        # Obtenir le choix de l'utilisateur
        user_item = self.get_user_item()
        
        # Obtenir le choix de l'ordinateur
        computer_item = self.get_computer_item()
        
        # Déterminer le résultat
        result = self.get_game_result(user_item, computer_item)
        
        # Afficher le résultat
        result_messages = {
            'gagné': f"Vous avez choisi {user_item}. L'ordinateur a choisi {computer_item}. Vous avez gagné !",
            'perdu': f"Vous avez choisi {user_item}. L'ordinateur a choisi {computer_item}. Vous avez perdu !",
            'égalité': f"Vous avez choisi {user_item}. L'ordinateur a choisi {computer_item}. Égalité !"
        }
        print(result_messages[result])
        
        # Retourner le résultat
        return result

# Exemple d'utilisation
game = Game()
resultat = game.play()  # Lance un jeu complet
```
