```python
import seaborn as sns

df = pd.read_csv('Student Mental health.csv')

print("Aperçu des données:")
print(df.head())
print("\nInformations sur les données:")
print(df.info())
print("\nStatistiques descriptives:")
print(df.describe())


df['Anxiety_Numeric'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})

# Calcul du pourcentage par genre
gender_anxiety = df.groupby('Choose your gender')['Anxiety_Numeric'].mean().reset_index()
gender_anxiety.rename(columns={'Anxiety_Numeric': 'Anxiety_Percentage'}, inplace=True)
gender_anxiety['Anxiety_Percentage'] = gender_anxiety['Anxiety_Percentage'] * 100

# Visualisation
plt.figure(figsize=(10, 6))
sns.barplot(x='Choose your gender', 
            y='Anxiety_Percentage', 
            data=gender_anxiety, 
            palette='Blues')

plt.title('Pourcentage d\'étudiants avec anxiété par genre', fontsize=14)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Pourcentage avec anxiété (%)', fontsize=12)
plt.ylim(0, 100)

# Ajout des étiquettes de pourcentage
for index, row in gender_anxiety.iterrows():
    plt.text(index, row['Anxiety_Percentage']+2, 
             f"{row['Anxiety_Percentage']:.1f}%", 
             ha='center')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

df = pd.read_csv('Student Mental health.csv')

print("Aperçu des données:")
print(df.head(10))
print("\nInformations sur les données:")
print(df.info())
print("\nStatistiques descriptives:")
print(df.describe())


df['Panic_Attack'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})


g = sns.lmplot(x='Age', y='Panic_Attack', data=df, logistic=True, height=6, aspect=1.5)



g.set(title="Relation entre l'âge et la probabilité d'attaques de panique\n",
      xlabel="Âge de l'étudiant",
      ylabel="Probabilité d'attaque de panique")
plt.show()
```
