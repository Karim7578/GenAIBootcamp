Load and Inspect Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
happiness_data = pd.read_csv('2019.csv')

# Display basic info
happiness_data.info
happiness_data.describe 
```

2 Data Preprocessing:

Address missing values and adjust data types as necessary.

```python
# missing values
print(happiness_data.isnull().sum())

# Define numerical columns (adjust according to actual CSV)
numerical_columns = ['Score', 'GDP per capita', 'Social support',
                   'Healthy life expectancy', 'Freedom to make life choices',
                   'Generosity', 'Perceptions of corruption']

# Filter to only existing columns
numerical_columns = [col for col in numerical_columns if col in happiness_data.columns]
print("\nNumerical columns used:", numerical_columns)

# Handle missing values by median imputation
for column in numerical_columns:
    median = happiness_data[column].median()
    happiness_data[column].fillna(median, inplace=True)
```

```python

# Define numerical columns (adjust according to actual CSV)
numerical_columns = ['Score', 'GDP per capita', 'Social support',
                   'Healthy life expectancy', 'Freedom to make life choices',
                   'Generosity', 'Perceptions of corruption']

# Filter to only existing columns
numerical_columns = [col for col in numerical_columns if col in happiness_data.columns]
print("\nNumerical columns used:", numerical_columns)

# Handle missing values by median imputation
for column in numerical_columns:
    median = happiness_data[column].median()
    happiness_data[column].fillna(median, inplace=True)

# Verify missing values after treatment
print("\n=== Missing Values After Treatment ===")
print(happiness_data.isnull().sum())

```



4 Regional Comparisons of GDP and Health:

    Create a Matplotlib subplot that compares â€˜GDP per Capitaâ€™ and â€˜Healthy Life Expectancyâ€™ across different regions.
    Use a bar plot for â€˜GDP per Capitaâ€™ and a line plot for â€˜Healthy Life Expectancyâ€™ on the same axes to observe how economic strength relates to health outcomes in different regions.



```python
# Objective: Investigate relationship between Social Support and Happiness Score
if 'Social support' in happiness_data.columns and 'Score' in happiness_data.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(happiness_data['Social support'], happiness_data['Score'], 
               alpha=0.6, color='purple')
    plt.title('Relationship Between Social Support and Happiness Score (2019)')
    plt.xlabel('Social Support (0-1)')
    plt.ylabel('Happiness Score (0-10)')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("\nCannot create plot: Missing required columns")

# Objective: Compare GDP and Health metrics across regions
if 'Region' in happiness_data.columns:
    regional_means = happiness_data.groupby('Region')[['GDP per capita', 
                                                     'Healthy life expectancy']].mean()
    
    # GDP Comparison (Bar Plot)
    plt.figure(figsize=(12, 6))
    regional_means['GDP per capita'].plot(kind='bar', 
                                        color='skyblue', 
                                        edgecolor='black')
    plt.title('Average GDP per Capita by Region (2019)')
    plt.xlabel('Region')
    plt.ylabel('GDP per Capita')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
    
    # Health Comparison (Bar Plot)
    plt.figure(figsize=(12, 6))
    regional_means['Healthy life expectancy'].plot(kind='bar', 
                                                 color='lightgreen', 
                                                 edgecolor='black')
    plt.title('Healthy Life Expectancy by Region (2019)')
    plt.xlabel('Region')
    plt.ylabel('Years')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
else:
    print("\nCannot create regional plots: 'Region' column missing")

# ----- 4.3 Correlation Matrix -----
plt.figure(figsize=(10, 8))
correlations = happiness_data[numerical_columns].corr()
sns.heatmap(correlations, 
           annot=True, 
           cmap='coolwarm', 
           center=0, 
           fmt=".2f",
           linewidths=0.5, 
           linecolor='white')
plt.title('Correlations Between Happiness Factors', pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
