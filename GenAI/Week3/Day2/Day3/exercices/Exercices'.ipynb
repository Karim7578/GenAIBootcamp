```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder

# Chargement des donnÃ©es
df = pd.read_csv('train.csv')

## 1. CrÃ©ation de nouvelles features
# Family Size = SibSp + Parch + 1 (la personne elle-mÃªme)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Extraction du titre depuis le nom (Mr, Mrs, Miss, etc.)
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Simplification des titres moins courants
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 
                                  'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df['Title'] = df['Title'].replace('Mlle', 'Miss')
df['Title'] = df['Title'].replace('Ms', 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')

## 2. Conversion des variables catÃ©gorielles
# Label Encoding pour le sexe
label_encoder = LabelEncoder()
df['Sex_encoded'] = label_encoder.fit_transform(df['Sex'])

# One-Hot Encoding pour les titres
title_dummies = pd.get_dummies(df['Title'], prefix='Title')
df = pd.concat([df, title_dummies], axis=1)

# One-Hot Encoding pour Embarked (port d'embarquement)
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, embarked_dummies], axis=1)

## 3. Normalisation/Standardisation des variables numÃ©riques
# Initialisation du scaler
scaler = StandardScaler()

# Standardisation de l'Ã¢ge et du tarif
df[['Age_scaled', 'Fare_scaled']] = scaler.fit_transform(df[['Age', 'Fare']])

# Affichage des rÃ©sultats
print("\nDataFrame aprÃ¨s transformations:")
print(df.head())

# VÃ©rification des nouvelles colonnes
print("\nColonnes crÃ©Ã©es:")
print(df[['FamilySize', 'Title', 'Sex_encoded', 'Age_scaled', 'Fare_scaled']].head())
print("\nColonnes one-hot encodÃ©es:")
print(title_dummies.head())
```

```python
## 1. Charger les donnÃ©es
df = pd.read_csv('train.csv')

## 2. Ã‰valuer l'Ã©chelle et distribution (version corrigÃ©e)
def assess_data(df, show_plots=True):
    # Statistiques descriptives
    print("Statistiques descriptives:")
    print(df.describe())
    
    if show_plots:
        # Une seule figure pour tous les histogrammes
        plt.figure(figsize=(12, 10))
        df.select_dtypes(include=['number']).hist()
        plt.tight_layout()
        plt.show()
    
    # Analyse dÃ©taillÃ©e
    print("\nAnalyse dÃ©taillÃ©e par colonne:")
    num_cols = df.select_dtypes(include=['number']).columns
    for col in num_cols:
        print(f"\n{col}:")
        print(f"Plage: {df[col].min()} Ã  {df[col].max()}")
        print(f"Moyenne: {df[col].mean():.2f} Â± {df[col].std():.2f}")
        print(f"Skewness: {df[col].skew():.2f}")

# Premier appel avec visualisation
assess_data(df)

## 3. PrÃ©traitement des donnÃ©es
wide_range_cols = ['Age', 'Fare']
bounded_cols = ['SibSp', 'Parch']
df_processed = df.copy()

## 4. Standardisation
scaler_standard = StandardScaler()
df_processed[wide_range_cols] = scaler_standard.fit_transform(df_processed[wide_range_cols])

## 5. Normalisation
scaler_minmax = MinMaxScaler()
df_processed[bounded_cols] = scaler_minmax.fit_transform(df_processed[bounded_cols])

## 6. VÃ©rification (sans rÃ©afficher les plots)
print("\nAprÃ¨s transformation:")
assess_data(df_processed, show_plots=False)  # DÃ©sactive les plots pour la vÃ©rification

print("\nDataFrame transformÃ©:")
print(df_processed.head())
```
