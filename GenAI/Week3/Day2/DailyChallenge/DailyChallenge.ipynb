
    Flip an image horizontally and then vertically using mirror and flip functions from ImageOps


```python
from PIL import Image, ImageOps

# Ouvrir l'image
image = Image.open('0001.png')  # Remplacez par votre chemin d'image

# Miroir horizontal (retournement gauche/droite)
flipped_horizontal = ImageOps.mirror(image)

# Miroir vertical (retournement haut/bas)  partir de l'image dj horizontale
flipped_both = ImageOps.flip(flipped_horizontal)

# Sauvegarder le rsultat
flipped_both.save("flipped_both_ways.png")

# Afficher le rsultat (optionnel)
flipped_both.show()
```
