```python
from scipy.ndimage import rotate
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from PIL import ImageOps
```


    To load an image and display it :


```python
original_image = Image.open('0001.png')

# Display the original image using matplotlib
plt.imshow(original_image)
```

Rotate an image by 90 degrees

```python
# Afficher l'image originale
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title("Image originale")

# Convertir l'image PIL en tableau NumPy (ncessaire pour scipy.ndimage.rotate)
image_array = np.array(original_image)

# Dfinir la fonction de rotation
def rotate_image_90_degrees(image):
    return rotate(image, 90, reshape=False, mode='nearest')

# Appliquer la rotation
rotated_image = rotate_image_90_degrees(image_array)

# Afficher l'image tourne
plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title("Image tourne de 90")

plt.show()
```


    Flip an image horizontally and then vertically using mirror and flip functions from ImageOps


```python
h_flipped = ImageOps.mirror(original_image)
v_flipped = ImageOps.flip(original_image)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
#flipped
axs[0].imshow(h_flipped)
axs[0].set_title('Flipped Horizontally')
axs[0].axis('off')
axs[1].imshow(v_flipped)
axs[1].set_title('Flipped Vertically')
axs[1].axis('off')
plt.show()
```

```python
width, height = original_image.size
zoomed = original_image.resize((int(width * 1.2), int(height * 1.2)))
#Zomm
plt.imshow(zoomed)
plt.title("Zoomed x1.2")
plt.axis('off')
plt.show()
```
