```python
# Importation des bibliothèques nécessaires (débutant friendly)
import pandas as pd
import matplotlib.pyplot as plt

# Correcting the filename
df = pd.read_csv('Time Americans Spend Sleeping.csv')

print(df.head())

print(df.info())

print(df.describe())

plt.figure(figsize=(10,5))

# Assuming 'Avg hrs per day sleeping' is a quantitative column
quant_col_name = 'Avg hrs per day sleeping'

if quant_col_name in df.columns:
    quant_col = df[quant_col_name]

    # Calculate statistics
    mean_val = quant_col.mean()
    median_val = quant_col.median()
    mode_val = quant_col.mode()[0]  # Taking first mode if multimodal

    print(f"Mean: {mean_val:.2f}")
    print(f"Median: {median_val:.2f}")
    print(f"Mode: {mode_val:.2f}\n")

    # Histogramme
    plt.subplot(1, 2, 1)
    df[quant_col_name].hist(bins=20, color='skyblue')
    plt.title(f"Distribution de {quant_col_name}")
    plt.xlabel(quant_col_name)
    plt.ylabel("Fréquence")
    plt.show()
else:
    print(f"Column '{quant_col_name}' not found in the DataFrame.")
```
